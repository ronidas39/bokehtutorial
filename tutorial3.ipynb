# tutorial 3,how to load & display plot from jupyter notebook

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure,show
```

```python
test_plot=figure()
test_plot.line([2,4,6,8],[10,20,30,40],line_width=4,line_color="red")
output_notebook()
```

```python
show(test_plot)
```
